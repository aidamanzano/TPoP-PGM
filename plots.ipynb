## Plots of general environment

```python
import numpy as np
import matplotlib.pyplot as plt
import simulator_functions as sim
import pandas as pd
```

```python
fig = plt.figure()


#fig.suptitle('Expected number of edges in Tree of depth 1, 6 neighbours per depth')
directory_pathfile = '/Users/amm3117/Desktop/TPoP PGM simulator/probabilistic_simulations/depth_1_neighbours_6/'
dataset = sim.full_csv(directory_pathfile)


simulations = pd.DataFrame(dataset)


z = 'Expected Number of Edges'
x = 'Probability of honest cars'
y = 'Probability of coerced cars'


df_temp = simulations[[x, y, z]].reset_index(drop=True)
df_temp = df_temp.groupby([x, y]).mean().reset_index()
#print(df_temp.sort_values([x, y]))

xlist = df_temp[x].values
ylist = df_temp[y].values
X, Y = np.meshgrid(xlist, ylist)
Z = np.zeros(X.shape)

#for each FP that equals 0, set it to a very small value below 1/1000 
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        z_val = df_temp[z][
            (df_temp[y] == Y[i, j]) & (df_temp[x] == X[i, j])
        ]
        if z_val.empty:
            Z[i, j] = None

        else:
            Z[i, j] = z_val
plt.contourf(X, Y, Z, vmin = 0.0, vmax = 6, cmap='RdYlGn', levels = 12)
    

plt.xlabel(x)
plt.ylabel(y)
plt.colorbar()
#fig.colorbar(cp)

ax = fig.get_figure()
```
